```python
import pickle
import networkx as nx
import pandas as pd
GraphDataPath = "../public/data.pickle"
DiGraphDataPath = "../public/dataDi.pickle"
# 读取数据
allFrmIndexDict = {}
dfs = []

with open(GraphDataPath, 'rb') as f:
    networkData = pickle.load(f)
    allGraph = {}
    for logId in list(networkData.keys()):
        
        logData = networkData[logId]
        nodes_num = []
        edges_num = []
        graphData = {}
        for frm in logData["frames"]:
            G = nx.Graph()
            G.add_nodes_from(logData[str(frm)]["nodes"])
            G.add_edges_from(logData[str(frm)]["edges"])
            nodes_num.append(len(logData[str(frm)]["nodes"]))
            edges_num.append(len(logData[str(frm)]["edges"]))
            graphData[str(frm)] = {'graph': G, 'egoInfos': logData[str(frm)]["ego"]}
#             allFrmIndexDict[idx] = {'logId':logId, 'frame':frm}
#             idx + 1
        allGraph[logId] = graphData
        subDf = pd.DataFrame({'nodes_num':nodes_num, 'edges_num':edges_num})
        subDf['logId'] = logId
        dfs.append(subDf)
logList = list(networkData.keys())
print(len(logList))
```

```python
graphStat = pd.concat(dfs)
```
